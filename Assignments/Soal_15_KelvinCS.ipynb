```python
import numpy as np
import pandas as pd

df = pd.read_html('http://digidb.io/digimon-list/')
df = df[0]
```

```python
df=df.set_index('#')
```

```python
len(df[(df['Type']=='Virus')&(df['HP']>df['HP'].mean())&(df['Atk']>100)&(df['Spd']>=df['Spd'].mean())])
```

```python
df.to_csv('Digimon.csv',index = False)
```

```python
df.groupby('Type').agg({'Digimon':'count'}).rename(columns={'Digimon':'Jumlah'})
```
